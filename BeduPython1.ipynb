```python
#renombrando los valores de los df
dict_trimestre = {220:"Abr-Jun", 320:"Jul-Sept",420:"Oct-Dic",121:'Ene-Mar'}
dict_genero = {0:"Sin informacion",1:"Hombre",2:"Mujer"}
#dict_edad = {0:"Sin informacion",98:"No especificado adultos",99:"No especificados menores"}
dict_PEA = {0:"Sin informacion",1:"PEA",2:"PNEA"} #Falta especificar el cero
dict_claspob = {0:"Sin informacion",1:"Poblacion ocupada",2:"Poblacion desocupada",3:"Disponible",4:"No disponibles"} #Falta definir el valor cero
dict_posocup = {0:"Sin informacion",1:"Subordinados y remunerados",2:"Empleadores",3:"Por cuenta propia",4:"Sin pago",5:"No especificado"} #Falta especificar el cero
dict_medico = {0:"Sin informacion",1:"Sin prestaciones",2:"Solo acceso a instituciones de salud",3:"Acceso a instituciones de salud y otras prestaciones",4:"No tiene acceso a instituciones de salud pero sí a otras prestaciones",5:"No especificado"} #Falta especificar el cero
dict_empleo = {0:"Sin informacion",1:"Informal",2:"Formal"} #Falta especificar el cero
```

```python
#renombrando las columnas
df_base_renamed["Trimestre_covid19"] = df_base_renamed["Trimestre_covid19"].map(dict_trimestre)
df_base_renamed["Genero"] = df_base_renamed["Genero"].map(dict_genero)
#df_base_renamed["Edad"] = df_base_renamed["Edad"].map(dict_edad)
df_base_renamed.loc[df_base_renamed["Edad"] == 98, 'Edad'] = "No especificado adultos"
df_base_renamed.loc[df_base_renamed["Edad"] == 99, 'Edad'] = "No especificados menores"
df_base_renamed["PEA_PNEA"] = df_base_renamed["PEA_PNEA"].map(dict_PEA)
df_base_renamed["Clasificacion_pob"] = df_base_renamed["Clasificacion_pob"].map(dict_claspob)
df_base_renamed["Posicion_ocupacion"] = df_base_renamed["Posicion_ocupacion"].map(dict_posocup)
df_base_renamed["Seguro_medico"] = df_base_renamed["Seguro_medico"].map(dict_medico)
df_base_renamed["Clasificacion_empleo"] = df_base_renamed["Clasificacion_empleo"].map(dict_empleo)
```
